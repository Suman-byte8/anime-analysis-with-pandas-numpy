**Function to extract the episode from Title**

```python
def extract_eps(txt):
  check = False
  data = ""
  for i in txt :
    if i == ')':
      check = False
      return data
    if check == True:
      data+=i
    if i == '(':
      check = True


df['Title'].apply(extract_eps).head()
```

**Added New Column for Episodes**

```python
df['Episodes'] = df['Title'].apply(extract_eps)
df.head()
```

**Extract the Timestamps from Title**

```python
df.loc[0]['Title']
```

```python
import re
from datetime import datetime

def extract_time(title_text):
    # Regex to find patterns like "Apr 2009 - Jul 2010" or "Jan 2021"
    # It looks for a sequence of 3 letters (month), space, 4 digits (year),
    # optionally followed by " - " and another month-year pattern.
    # The entire pattern is captured after a closing parenthesis and optional spaces.
    match = re.search(r'\)\s*([A-Za-z]{3}\s*\d{4}(?:\s*-\s*[A-Za-z]{3}\s*\d{4})?)', title_text)
    if match:
        return match.group(1).strip()
    return None
```

```python
df['Timestamp'] = df['Title'].apply(extract_time)
```

**Anime with Longest Duration**

```python
def calculate_duration_months(timestamp_str):
    if not isinstance(timestamp_str, str) or timestamp_str is None:
        return 0

    parts = timestamp_str.split(' - ')

    try:
        if len(parts) == 2:
            start_dt = datetime.strptime(parts[0].strip(), '%b %Y')
            end_dt = datetime.strptime(parts[1].strip(), '%b %Y')
            # Calculate duration in months, inclusive of both start and end months
            duration_months = (end_dt.year - start_dt.year) * 12 + (end_dt.month - start_dt.month) + 1
            return max(0, duration_months) # Ensure duration is not negative
        elif len(parts) == 1:
            # If only one date is present, consider it as a 1-month duration (e.g., for movies/specials)
            return 1
        else:
            return 0 # Handle unexpected format
    except ValueError:
        return 0 # Handle any parsing errors

# Re-apply extract_time to ensure the latest version is used for Timestamp column
df['Timestamp'] = df['Title'].apply(extract_time)

df['Duration (Months)'] = df['Timestamp'].apply(calculate_duration_months)

longest_anime = df.loc[df['Duration (Months)'].idxmax()]
display(longest_anime[['Title', 'Timestamp', 'Duration (Months)']])
```
