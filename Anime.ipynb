**Import Numpy & Pandas**

```python
import numpy as np
import pandas as pd
```

**Load CSV File**

```python
df = pd.read_csv('/content/anime.csv')
display(df.head())
```

**Seeing the Title Example from First Row**

```python
df.iloc[0]['Title']
```

**Function to extract the episode from Title**

```python
def extract_eps(txt):
  check = False
  data = ""
  for i in txt :
    if i == ')':
      check = False
      return data
    if check == True:
      data+=i
    if i == '(':
      check = True


df['Title'].apply(extract_eps).head()
```

**Added New Column for Episodes**

```python
df['Episodes'] = df['Title'].apply(extract_eps)
df.head()
```

**Removing eps from Episodes Column**

```python
df['Episodes'] = df['Episodes'].str.replace(' eps',"")
```

```python
df.head()
```

**Converting Episodes dataType str to int**

```python
type(df.iloc[0]['Episodes'])
```

```python
df['Episodes'] = df['Episodes'].astype(int)
```
